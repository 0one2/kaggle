```python
train = pd.read_csv("../input/train.csv", parse_dates = ["datetime"])
test = pd.read_csv("../input/test.csv", parse_dates = ["datetime"])
```

```python
train["year"] = train["datetime"].dt.year
train["hour"] = train["datetime"].dt.hour
train["dayofweek"] = train["datetime"].dt.dayofweek

test["year"] = test["datetime"].dt.year
test["hour"] = test["datetime"].dt.hour
test["dayofweek"] = test["datetime"].dt.dayofweek
```

```python
y_train = train["count"]
y_train = np.log1p(y_train)
```

```python
y_casual = np.log1p(train.casual)
y_registered = np.log1p(train.registered)
```

```python
X_train = train.drop(["datetime", "windspeed", "casual", "registered", "count"], 1)
X_test = test.drop(["datetime", "windspeed"], 1)
```

```python
import lightgbm as lgb
```

```python
hyperparameters = {'boosting_type': 'gbdt', 'colsample_bytree': 0.7250136792694301, 'is_unbalance': False, 'learning_rate': 0.013227664889528229,
                   'min_child_samples': 20, 'num_leaves': 56, 'reg_alpha': 0.7543896477745794, 'reg_lambda': 0.07152751159655985, 
                   'subsample_for_bin': 240000, 'subsample': 0.5233384321711397, 'n_estimators': 1093}
```

```python
model = lgb.LGBMRegressor(**hyperparameters)
```

```python
model.fit(X_train, y_casual)
preds1 = model.predict(X_test)
```

```python
hyperparameters = {'boosting_type': 'gbdt', 'colsample_bytree': 0.6399481302431778, 'is_unbalance': False, 'learning_rate': 0.011762357512839758,
                   'min_child_samples': 20, 'num_leaves': 30, 'reg_alpha': 0.3519969279711128, 'reg_lambda': 0.587719028796242, 'subsample_for_bin': 120000,
                   'subsample': 0.9161483473338935, 'n_estimators': 2166}
```

```python
model = lgb.LGBMRegressor(**hyperparameters)
```

```python
model.fit(X_train, y_registered)
preds2 = model.predict(X_test)
```

```python
sub = pd.read_csv("../input/sampleSubmission.csv") 
preds = np.expm1(preds1) + np.expm1(preds2)
sub["count"] = preds
sub.to_csv("lgbm_casual_registered.csv", index=False)
```
