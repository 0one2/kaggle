```python
X_train = train.drop(["datetime", "windspeed", "casual", "registered", "count"], 1)
X_test = test.drop(["datetime", "windspeed"], 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model1  = RandomForestRegressor(n_estimators=110, n_jobs = -1, random_state=43)
model2  = RandomForestRegressor(n_estimators=110, n_jobs = -1, random_state=43)
```

```python
model1.fit(X_train, y_casual)
model2.fit(X_train,y_registered)
pred1 = model1.predict(X_test)
pred2 = model2.predict(X_test)
```

```python
sub = pd.read_csv("../input/sampleSubmission.csv") 
sub["count"] = pred1+ pred2
sub.to_csv("result.csv", index=False)
```
