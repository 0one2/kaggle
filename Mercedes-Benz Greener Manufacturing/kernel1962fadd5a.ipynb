```python
#93 107 233 235 268 289 290 293 297 330 347 
cata_cols = ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']

alldata = pd.concat([df_train, df_test], sort=False)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
alldata[cata_cols] = alldata[cata_cols].apply(le.fit_transform)
```

```python
alldata.head()
```

```python
alldata.head()
```

```python
alldata.drop(['X93','X107','X233','X235','X268','X290','X293','X297','X330','X347'],axis = 1,inplace = True)
```

```python
df_train = alldata.iloc[:len(df_train), :]
df_test = alldata.iloc[len(df_train):, :]
```

```python
y = df_train['y']
X = df_train.drop(['y'], axis=1)
X_test = df_test.drop(['y'], axis=1)
```

```python
from catboost import CatBoostRegressor
```

```python
model = CatBoostRegressor(random_seed=42,depth= 4)
```

```python
model.fit(X, y, cat_features= [1,2,3,4,5,6,7,8], )
```

```python
cat_preds = model.predict(X_test)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=41)
models = []
best_iter = 0
_y_predict = np.zeros(X.shape[0])
```

```python
for train_index, valid_index in kf.split(X):
    X_train, X_valid = X.iloc[train_index, :], X.iloc[valid_index, :]
    y_train, y_valid = y.values[train_index], y.values[valid_index]
    model = xgb.XGBRegressor(max_depth=4,seed=1,)
    model.fit(X_train, y_train,  early_stopping_rounds=75, verbose=False, eval_set=[(X_train, y_train), (X_valid, y_valid)], eval_metric="rmse")
    best_iter += model.best_iteration
    _y_predict[valid_index] = model.predict(X_valid)
```

```python
best_iter
```

```python
best_iter //= 5
model = xgb.XGBRegressor(max_depth=4, n_estimators=best_iter,seed=1,)
model.fit(X, y)
```

```python
xg_preds = model.predict(X_test)
```

```python
preds = 0.7*cat_preds + 0.3*xg_preds
submit = pd.DataFrame({'ID': df_test['ID'], 'y': preds})
submit.to_csv('benz-cat.csv', index=False)
```
